```python
# Visualization: Create a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_pivot_radiation, fmt=".1e", cmap='YlGnBu', linewidths=0.5,cbar=True)

# Adding labels and title
plt.title('Heatmap of Sum of Radiation per Month for Each Year')
plt.xlabel('Year')
plt.ylabel('Month')
plt.xticks(rotation=45)  # Rotate year labels for better visibility
plt.yticks(ticks=range(12), labels=[
    'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'
], rotation=0)  # Setting month labels

# Show the plot
plt.tight_layout()
plt.show()
```

### Skewness

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Create a box plot grouped by 'Region'
df_radiation_visualize.boxplot(column='Sum Radiation', by='region', grid=False, vert = False)

# Add labels and title
plt.title('Box Plot of Radiation in (KJ/m2) by Region')
plt.suptitle('')  # Suppress default subtitle
plt.xlabel('Radiation in (KJ/m2)')
plt.ylabel('Region')

# Display the plot
plt.show()
```

```python
skewness = df_radiation_visualize['Radiation'].skew()
#show the Inflow of the skewness
print(f"Skewness of 'Radiation' column: {skewness}")
# Determine and print if the skewness is left or right
if skewness > 0:
    print("The data is right-skewed (positive skew).")
elif skewness < 0:
    print("The data is left-skewed (negative skew).")
else:
    print("The data is symmetric (no skew).")
```

```python
skewness = df_wind_visualize['Average Wind Speed (m/s)'].skew()
#show the Inflow of the skewness
print(f"Skewness of 'Average Wind Speed (m/s)' column: {skewness}")
# Determine and print if the skewness is left or right
if skewness > 0:
    print("The data is right-skewed (positive skew).")
elif skewness < 0:
    print("The data is left-skewed (negative skew).")
else:
    print("The data is symmetric (no skew).")
```

This means, we will need to use a function on our data. 
For right skewed data: 
- square root
- logarithm
